```python
"Coroutine Method Faster"
import aiohttp
import asyncio
import pandas as pd
from bs4 import BeautifulSoup

Tabela = []

url = "https://www.impic.pt/impic/pt-pt/consultar/empresas-titulares-de-licenca-de-mediacao-imobiliaria"
# Tabela
id_type = str(8) 
id_object = str(25)#e o que define qual tabela é independentemente de url
# Paginas
pag_inicial = 1
pag_final = 473
# cookie
PHPSESSID = str("k1pam5fjprdrhg90ud5hj5ega4")

async def fetch(session, url, data):
    async with session.post(url, data=data, verify_ssl=False) as response:
        return await response.text()

async def table(x, y, url=url):
    """
    Function to extract a table from a given page with given parameters x and y (pages).
    """
    cookies = {
        'PHPSESSID': PHPSESSID,
    }
    
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:87.0) Gecko/20100101 Firefox/87.0',
        'Accept': '*/*',
        'Accept-Language': 'en-US,en;q=0.8,pt-PT;q=0.5,pt;q=0.3',
        'Content-Type': 'application/x-www-form-urlencoded; charset=UTF-8',
        'X-Requested-With': 'XMLHttpRequest',
        'Origin': 'http://www.impic.pt',
        'DNT': '1',
        'Connection': 'keep-alive',
        'Referer': url,
        'Pragma': 'no-cache',
        'Cache-Control': 'no-cache',
    }
    
    data = {
      'id_type': id_type,
      'id_object': id_object,
      'pesquisar': 'true',
      'loadTable': '{}'.format(x),
      'pageSearch': '{}'.format(y)
    }

    async with aiohttp.ClientSession() as session:
        async with session.post('https://www.impic.pt/impic/ajax/call/impic_api/consultar/ajax/43', headers=headers, cookies=cookies, data=data,verify_ssl=False) as response:
            html = await response.text()#could maybe do a print here to show status
            soup = BeautifulSoup(html, "html5lib")
            
            table = soup.table
            if not table:
                return
            table_row = table.find_all("tr")
            for tr in table_row:
                td = tr.find_all("td")
                row = [i.text for i in td]
                Tabela.append(row)


async def main():
    """
    Main function to iterate through pages and extract tables.
    """
    tasks = []
    for i in range(pag_inicial, pag_final+1):
        # i=[1,pagina final]
        task = asyncio.ensure_future(table(1, i))
        #print("page:{} done".format(i))
        tasks.append(task)
    await asyncio.gather(*tasks)

await main()


# to excell
# df.to_excel("Impic.xlsx")
df = pd.DataFrame(Tabela)
df.to_excel(r"C:\Users\joaot\Desktop\Impic.xlsx")
```
